### Exercise 2: Get the Total profit of all months and show a line plot with the following Style properties:
- Line Style dotted and Line color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

The line plot graph should look like this:


```python
# Write your code here
x = data.month_number
y = data.total_profit
plt.figure(figsize=(10,7))
plt.plot(x,y, color='red', linewidth=3, marker='o', markerfacecolor='black', linestyle='--', label="total_profit")
plt.title("Comapny sales data of last year")
plt.legend(loc='lower right')
plt.xlabel("Month number")
plt.xticks(tuple(x), rotation=0)
plt.ylabel("Profit")
plt.show()
```

### Exercise 8: Calculate the total sale data for last year for each product and show it using a Pie chart.
Note: In the Pie chart you should display the Number of units sold per year for each product in percentage.

The Pie chart should look like this:


```python
# Write your code here
features = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
sum_list = []
for feature in features:
    sum_list.append(data[feature].sum())
print(sum_list)

plt.figure(figsize=(12,10))
plt.axis("equal")
plt.pie(sum_list, labels=features, autopct='%1.1f%%', textprops={'color':"blue",'fontsize': 20})
plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
```
