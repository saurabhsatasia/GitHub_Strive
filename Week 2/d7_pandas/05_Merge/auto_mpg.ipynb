## MPF Cars

```python
import pandas as pd
import numpy as np

url1 = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv'
url2 = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv'

car1 = pd.read_csv(url1)
car2 = pd.read_csv(url2)

car1.head(2)
```

```python
car2.head(2)
```

```python
car1.columns
```

```python
car1 = car1.drop(['Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13'], axis=1)
car1.head(2)
```

#### 1. Join cars1 and cars2 into a single DataFrame called cars

```python
cars = car1.append(car2)
cars
```

#### 2. There is a column missing, called owners. Create a random number Series from 15,000 to 73,000.

```python
nr_owners = np.random.randint(15000, high=73001, size=398, dtype='l')
nr_owners
```
